In this section, I displayed all the uniques value of the Reviews column using the unique() method of pandas.

```python
Reviews = []

for x in df.Reviews:
    x= x.replace('M', '00')
    Reviews.append(x)

Reviews = list(map(float, Reviews))
df['reviews'] = Reviews
```

In this section, we created a loop that convert all the value of column Reviews of our dataset into float format using replace() function of pandas. And We put all these values in new format in a new columns called reviews.

```python
#Installs
df.Installs.unique()
```

In this section, we displayed all the unique values in the Installs coulumns of our dataset using unique() function of pandas.

```python
df['Installs'].replace(to_replace=['0', 'Free'], value=['0+', '0+'], inplace=True)
Installs = []

for x in df.Installs:
    x= x.replace(',', '')
    Installs.append(x[:-1])
    
Installs = list(map(float, Installs))
df['Installs'] = Installs
```

In this section, we created a loop that remove all the commas inside the value of column Installs of our dataset using the replace method of pandas. And We update the column Installs with values in new format.

```python
#Total reviews
plt.figure(figsize=(12,12))
sum_inst = df.groupby(['Category'])['Installs'].sum().sort_values(ascending=False)
sns.barplot(x=sum_inst, y=sum_inst.index, data=df)
```

In this section, We used seanborn and compilation with matplotlib to visual our figure. We ploted using the method of matplotlib library, the graphique of sum of Installs values of each unique elements of the category column. We represent the level of installs values of those category by bars using the method barplot() of seaborn library of python.

```python
#Mean reviews
plt.figure(figsize=(12,12))
sum_inst = df.groupby(['Category'])['Installs'].mean().sort_values(ascending=False)
sns.barplot(x=sum_inst, y=sum_inst.index, data=df)
```

In this section, We used seanborn and compilation with matplotlib to visual our figure. We ploted using the method of matplotlib library, the graphique of mean of Installs values of each unique elements of the category column. We represent the level of installs values of those category by bars using the method barplot() of seaborn library of python.

```python
#Type for category
df.Type.unique()

#There is 0 and null value, let's change them to free.
df['Type'].replace(to_replace=['0'], value=['Free'], inplace=True)
df['Type'].fillna('Free', inplace=True)
print(df.groupby('Category')['Type'].value_counts())

#visualize
Type_cat = df.groupby('Category')['Type'].value_counts().unstack().plot.barh(figsize=(10, 20), width=0.7)
plt.show()
```

In this last section, We wrote the code first, to get all the unique value of the column Type of our dataset using unique function of pandas. Secondly we wrote the code to replace all the null values or the value O in the Type column with the word "Free" using the replace function() of pandas. Thirdly, we wrote the code to group each categogry using grouby method and we display them along with their types and the numbers times that type is repeated for that category. by example for the category weather, there is 74 for the TWype Free and 8 for the Type Paid. Lastly we visualise our result using matplotlib method, we used barh() to plot our data in form of progress bar.
